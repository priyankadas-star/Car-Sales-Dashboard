```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Unit sold"].sum()
```

```python
df["Price"].mean()
```

```python
df.groupby("Brand")["Unit sold"].sum().sort_values(ascending=False).head(10)
```

```python
brand_sales = df.groupby("Brand")["Unit sold"].sum()

brand_sales.plot(kind="bar")
plt.title("Brand-wise Sales")
plt.show()
```

```python
vehicle_sales = df.groupby("Vehicle type")["Unit sold"].sum()

plt.pie(vehicle_sales, labels=vehicle_sales.index, autopct='%1.1f%%')
plt.show()
```

```python
plt.scatter(df["Price"], df["Horsepower"])
plt.xlabel("Price")
plt.ylabel("Horsepower")
plt.show()
```
